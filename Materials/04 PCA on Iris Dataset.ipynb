```python
#Load the data
iris = load_iris()
X=iris.data
Y = iris.target
```

```python
Y
```

```python
# Center the data and get rid of the units
X=(X-X.mean(axis=0))/X.std(axis=0)
```

```python
# Find top 2 principal componets with PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
#Project the data into a two dimensional space
P=pca.transform(X)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.scatter(P[:,0], P[:,1], c=iris.target)
```
