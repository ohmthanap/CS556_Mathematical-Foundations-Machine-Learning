```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
from  numpy.linalg import matrix_rank as rank
```

```python
def image_show(X, title):
    img = plt.imshow(X)
    img.set_cmap('gray')
    plt.axis("off")
    plt.title(title)
    plt.show()  
def rbg_gray(filename):
    A = imread(filename)
    return np.mean(A, -1)
```

```python
img = rbg_gray("flower.jpg")
image_show(img, title = "Original Image")
#Show the image
```

```python
img.shape
```

```python
#Perform SVD Decompostion
U, S, VT = np.linalg.svd(img)
S = np.diag(S)
```

```python
rec_image = U@S@VT
```

```python
image_show(rec_image, 'Reconstructed Image')
```

```python
#Show different reduced version of the image
for r in [rank(img), 100, 50, 25, 5, 1]:
    Xapprox = U[:, :r]@S[:r, :r]@VT[:r, :]
    image_show(Xapprox, f'Rank r={r}')
```
